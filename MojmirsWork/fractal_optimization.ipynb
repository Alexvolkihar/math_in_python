# Fractal optimization

The purpose of this notebook is to try to speed up fractal generation of mandelbrot set as much as possible. I got inspiration from this code in [Cython](https://github.com/cython/cython/wiki/examples-mandelbrot) and tried to write a code which would run as fast using [pypy](http://pypy.org/).

*Note: The scripts below are not supposed to be run in iPython, use pypy instead*

## PyPy

From all the things I tried, those are the ones that had the greatest effect on speedup:

* Image is not being created with `putpixel` method, but rather buffered into an array.
* Coloring is not done continously, but a palette of colors is pre-generated and we pick colors from it.
* Type `complex` was replaced with tuple of numbers to avoid expensive square root when calculating complex number size.
* Slicing of numpy arrays is replaced by assignment (see code).

Although `pypy` should run without any code modifications, I got it working with decent speedup only after several hours due to various bugs and unexpected problems. It took at least 10x more time than writing the python code itself. Applying `pypy` on an original code `fractals.py` was as fast as regular python.

## Parallelization
I tried to parallelize it using multiprocessing module by splitting images into blocks which are then processed separately and concatenated at last. Unfortunately, PyPy didn't generate an image due to problems with array pickling.

## Pallete of colors

```python
from __future__ import division
import Image
import colorsys

colscale = lambda c: (int(c[0] * 255), int(c[1] * 255), int(c[2] * 255))

def linspace(start, stop, n):
    if n == 1:
        yield stop
        return
    h = (stop - start) / float(n - 1)
    for i in range(n):
        yield start + h * i


def palette(n=100, rng=3, system='RGB', scaled=False):
    h = map(lambda x: x % 1, linspace(0 + 0.6, 3 + 0.6, n))
    s = [1.] * n
    v = map(lambda x: x**0.2, linspace(0, 1, n))
    if system == 'RGB':
        cols = [colorsys.hsv_to_rgb(*x) for x in zip(h, s, v)]
    else:
        cols = zip(h, s, v)

    if scaled:
        cols = map(colscale, cols)

    return cols
```

## PyPy script

```python
try:
    import numpypy as np
    def linspace(start, stop, n):
        if n == 1:
            yield stop
            return
        h = (stop - start) / (n - 1)
        for i in range(n):
            yield start + h * i
except ImportError:
    import numpy as np
    linspace = np.linspace

import Image
import time

PALETTE = palette(100, rng=2, system='RGB', scaled=True)

def toimage(arr):
    data = np.asarray(arr)
    shape = list(data.shape)
    shape = (shape[1],shape[0])
    image = Image.fromstring('RGB', shape, data.tostring())
    return image


def mandelbrot(re, im, max_iter=500, escape=4.):
    z_im, z_re = 0., 0.
    i = 0
    while i < max_iter:
        z_re, z_im = (z_re * z_re - z_im * z_im + re, 2 * z_re * z_im + im)
        if z_re * z_re + z_im * z_im >= escape:
            return i
        i += 1
    return -1


def complex_image(size, center=(0,0), d=2):

    byte_array = (np.zeros((size, size, 3))).astype('uint8')

    palette_len = len(PALETTE)

    for i, a in enumerate(linspace(center[0] - d, center[0] + d, size)):
        for j, b in enumerate(linspace(center[1] - d, center[1] + d, size)):
            k = mandelbrot(a, b)

            if k != -1:
                byte_array[i, j, 0] = PALETTE[k % palette_len][0]
                byte_array[i, j, 1] = PALETTE[k % palette_len][1]
                byte_array[i, j, 2] = PALETTE[k % palette_len][2]

    im = toimage(byte_array)
    return im


if __name__ == '__main__':
    start = time.time()
    im = complex_image(500, center=(-0.75, -0.75), d=0.75)
    print "it took %f seconds to run"%( time.time() - start)

    im.show()
    ##im.save('mandel_fast.bmp')
else:
    import cProfile, pstats

    stats = cProfile.runctx('complex_image(500, center=(-0.75, -0.75), d=0.75)',
        globals(), locals(), 'stats')

    p = pstats.Stats('stats')
    p.strip_dirs().sort_stats('time').print_stats()
```

## PyPy multiprocessing

```python
from __future__ import division

try:
    import numpypy as np
    def linspace(start, stop, n):
        h = (stop - start) / float(n - 1)
        return [start + h * i for i in range(n)]

except ImportError:
    import numpy as np
    linspace = np.linspace

import Image
import time

import multiprocessing

PALETTE = palette(100, rng=2, system='RGB', scaled=True)

def toimage(arr):
    data = np.asarray(arr)
    shape = list(data.shape)
    shape = (shape[1],shape[0])
    image = Image.fromstring('RGB', shape, data.tostring())
    return image


def mandelbrot(re, im, max_iter=500, escape=4.):
    z_im, z_re = 0., 0.
    i = 0
    while i < max_iter:
        z_re, z_im = (z_re * z_re - z_im * z_im + re, 2 * z_re * z_im + im)
        if z_re * z_re + z_im * z_im >= escape:
            return i
        i += 1
    return -1


def complex_image_process(queue, size=(500,250), center=(-0.75, -0.75), d=0.75):
    height, width = size
    d_height, d_width = d, d * width / height
    byte_array = (np.zeros((height, width, 3))).astype('uint8')
    palette_len = len(PALETTE)

    for i, a in enumerate(linspace(center[1] - d_height, center[1] + d_height, height)):
        for j, b in enumerate(linspace(center[0] - d_width, center[0] + d_width, width)):
            k = mandelbrot(a, b)

            if k != -1:
                byte_array[i, j, 0] = PALETTE[k % palette_len][0]
                byte_array[i, j, 1] = PALETTE[k % palette_len][1]
                byte_array[i, j, 2] = PALETTE[k % palette_len][2]

    #queue.put(byte_array)


def complex_image(size=500, center=(-0.75, -0.75), d=0.75, processes=2):
    assert size % processes == 0, 'Processes must divide image size.'

    jobs = []

    queue = multiprocessing.Queue()

    for i in range(processes):
        re = center[0] - d + d / processes * (2*i + 1)
        kwargs = {'size': (size, int(size / processes)),
                  'center': (re, center[1]),
                  'd': d,
                  'queue': queue}
        p = multiprocessing.Process(target=complex_image_process, kwargs=kwargs)
        jobs.append(p)
        p.start()

    for j in jobs:
        j.join()

    block_list = []
    while not queue.empty():
        block_list.append(queue.get())

    if block_list:
        b = np.hstack(block_list)
        im = toimage(b)
        return im
    else:
        return None


if __name__ == '__main__':
    start = time.time()

    im = complex_image(500, center=(-0.75, -0.75), d=0.75, processes=2)

    print "it took %f seconds to run" % ( time.time() - start)

    if im:
        im.show()

else:
    import cProfile, pstats

    stats = cProfile.runctx('complex_image(500, center=(-0.75, -0.75), d=0.75)',
        globals(), locals(), 'stats')

    p = pstats.Stats('stats')
    p.strip_dirs().sort_stats('time').print_stats()
```
