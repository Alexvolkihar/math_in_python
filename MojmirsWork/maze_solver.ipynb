# Maze solver

We are given `.txt` file with maze representation like this one:

```
012F
1233
1313
S132
```

where numbers represent colors and S, resp. F, stands for start, resp. finish. The task is to find a way from start to finish which steps on each color equal number of times.

### Solution
To find a solution, we will use backtracking - use DFS for all neighboring tiles until you get to the situation when you have more tiles of one color than minimum of all colors from all tiles. In that case go back and try another route.

````python
import svgfig as svg
from collections import defaultdict


class Maze(list):
    start = None
    finish = None

    def colors(self):
        """ Return dict of colors and their numbers. """
        cols = defaultdict(int)
        for row in self:
            for c in row:
                if c != 0:
                    cols[c] += 1
        return cols


def load_maze(filename):
    """ Load maze from txt description and return matrix (list of lists). """
    with open(filename) as f:
        maze = Maze([])
        for i, line in enumerate(f):
            row = []
            for j, c in enumerate(line[:-1]):
                if c == 'S':
                    maze.start = (i, j)
                    c = 0
                elif c == 'F':
                    maze.finish = (i, j)
                    c = 0
                row.append(int(c))
            maze.append(row)
    return maze


def maze_picture(maze, path, **kwargs):
    """ Create .svg figure with maze. """
    COLOR_LIST = ['white', 'blue', 'green', 'yellow', 'red', 'orange', 'black']
    R = 100 / len(maze)

    s = svg.Fig().SVG()
    # draw tiles
    for i, row in enumerate(maze):
        for j, color in enumerate(row):
            s.append(svg.SVG("rect", x=j*R, y=i*R, width=R, height=R,
                               fill=COLOR_LIST[color]))

    # draw path
    trans = lambda (x, y): (int(x * R + 0.5 * R), int(y * R + 0.5 * R))
    coor_path = map(trans, path)
    for (y1,x1), (y2,x2) in zip(coor_path[:-1], coor_path[1:]):
        s.append(svg.SVG('line', x1=x1, y1=y1, x2=x2, y2=y2, stroke=COLOR_LIST[-1]))

    # start and end point
    for x,y in [maze.start, maze.finish]:
        cx, cy = trans((x, y))
        s.append(svg.SVG('circle', cx=cx, cy=cy, r=2, fill='red'))

    c = svg.canvas(s, **kwargs)
    return c


def dfs_maze(maze, path, ncolors, min_color):
    x, y = path[-1]

    if (x, y) == maze.finish and all(e == ncolors.values()[0] for e in ncolors.values()):
        return path

    # search through other vertices
    for i, j in [(-1,0), (1,0), (0,1), (0,-1)]:
        new_x = x + i
        new_y = y + j

        # crossing path
        if (new_x, new_y) in path:
            continue

        # out of bounds
        if not (0 <= new_x < len(maze) and 0 <= new_y < len(maze)):
            continue

        c = maze[new_x][new_y]
        new_ncolors = ncolors.copy()
        if c != 0:
            new_ncolors[c] += 1
            # too many of one color
            if new_ncolors[c] > min_color:
                continue

        new_path = path[:]  # copy path
        new_path.append((new_x, new_y))

        new_path = dfs_maze(maze, new_path, new_ncolors, min_color)
        if new_path:
            return new_path
    return None



def solve_maze(maze):
    # init
    path = [maze.start]
    ncolors = {c: 0 for c in maze.colors().keys()}
    min_color = min(maze.colors().values())

    # run DFS with backtracking
    path = dfs_maze(maze, path, ncolors, min_color)
    return path
````

````python
from IPython.display import SVG
maze = load_maze('data/maze1.txt')
path = solve_maze(maze)
s = maze_picture(maze, path)
s.save('img/maze-solver-1.svg')
SVG('img/maze-solver-1.svg')
````

````python
maze = load_maze('data/maze2.txt')
path = solve_maze(maze)
s = maze_picture(maze, path)
s.save('img/maze-solver-2.svg')
SVG('img/maze-solver-2.svg')
````
