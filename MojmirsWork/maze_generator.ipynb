# Maze generator

## Square maze

```python
from __future__ import division
import random
import svgfig as svg

class Maze(object):
    """ Return list of tuples (pos, new_pos) which represent
    line between two cells. """

    def __init__(self, n):
        self.n = n
        self.visited = [[False for _ in range(n)] for _ in range(n)]
        self.visited[0][0] = True
        self.lines = self.dfs((0,0), set())


    def dfs(self, pos, lines):
        x, y = pos
        dirs = [(-1,0), (1,0), (0,1), (0,-1)]
        random.shuffle(dirs)
        for i, j in dirs:
            new_x, new_y = x + i, y + j

            # out of bounds
            if not (0 <= new_x < self.n and 0 <= new_y < self.n):
                continue

            # cycle?
            if self.visited[new_x][new_y]:
                continue

            # go on
            self.visited[new_x][new_y] = True
            lines.add(tuple(sorted([(x, y), (new_x, new_y)])))
            self.dfs((new_x, new_y), lines)
        return lines

    def picture(self, skeleton=False):
        r = 100 / self.n
        s = svg.Fig().SVG()

        if skeleton:
            for (x,y), (new_x, new_y) in self.lines:
                s.append(svg.SVG('line', x1=x*r+0.5*r,
                                         y1=y*r+0.5*r,
                                         x2=new_x*r+0.5*r,
                                         y2=new_y*r+0.5*r, stroke='red'))

        for i in range(self.n+1):    # rows
            for j in range(self.n+1):    # columns
                # horizontal line
                if ((i,j-1), (i, j)) not in self.lines:
                    s.append(svg.SVG('line', x1=i*r, y1=j*r, x2=(i+1)*r, y2=j*r))
                # vertical line
                if ((i-1,j), (i, j)) not in self.lines:
                    s.append(svg.SVG('line', x1=i*r, y1=j*r, x2=i*r, y2=(j+1)*r))

        c = svg.canvas(s)
        return c
```

```python
from IPython.display import SVG
Maze(20).picture().save('img/square.svg')
SVG('img/square.svg')
```

## Triangle maze

To make the algorithm simpler, triangle is first represented as follows:

```python
SVG('img/triangle_design.svg')
```

This representation has an advantage in that we can index triangles with integer numbers as (row, column). First, I will find the spanning tree and save it as a set of vertices and then just create the walls. To get the maze into the final shape of equilateral triangle, affine transformation is applied on it.

```python
from math import sqrt
import random

class Vertex(object):

    def __init__(self, row, column):
        self.neighbors = set()
        self.row = row
        self.column = column

    def edge_to(self, vertex):
        self.neighbors.add(vertex)
        vertex.neighbors.add(self)

    def __repr__(self):
        return str((self.row, self.column))

    def __eq__(self, other):
        if isinstance(other, Vertex):
            return self.row == other.row and self.column == other.column
        elif isinstance(other, (list, tuple)):
            return (self.row, self.column) == tuple(other)

    def __ne__(self, other):
        return not self.__eq__(other)

    def __hash__(self):
        return (self.row, self.column).__hash__()


class TriangleMaze(object):

    def __init__(self, n):
        self.n = n
        self.vertices = self.create_vertices(n)

        init_vertex = random.sample(self.vertices, 1)[0]
        visited = set([init_vertex])
        self.path = self.dfs(init_vertex, set(), visited)


    def dfs(self, vertex, path, visited):
        neighbors = list(vertex.neighbors)
        random.shuffle(neighbors)
        for v in neighbors:
            if v in visited:
                continue

            # visit new vertex
            visited.add(v)
            path.add((vertex, v))
            path.add((v, vertex))    # adding both permutations is inefficient,
                                     # but simplifies later code
            self.dfs(v, path, visited)
        return path


    def create_vertices(self, n):
        vertices = {}
        for column in range(n):
            for row in range(2 * (n - column) - 1):
                v = Vertex(row, column)
                vertices[(row, column)] = v

                # edge down
                if row != 0:
                    v.edge_to(vertices[row - 1, column])

                # edge left
                if column != 0 and row % 2 == 0:
                    v.edge_to(vertices[row + 1, column - 1])

        return vertices.values()


    def draw(self):
        n = self.n
        path = self.path

        walls = []
        for column in range(n):
            for row in range(2 * (n - column) - 1):
                if row % 2 == 0:
                    r = int(row / 2)    # normalize indexing
                    # wall above
                    if ((row-1, column), (row, column)) not in path:
                        walls.append(((column, r), (column + 1, r)))

                    # wall on the left
                    if ((row+1, column-1), (row, column)) not in path:
                        walls.append(((column, r), (column, r + 1)))

                    # diag wall
                    if ((row, column), (row+1, column)) not in path:
                        walls.append(((column, r + 1), (column + 1, r)))

        # affine transformation
        s = svg.Fig().SVG()
        scale = lambda (x,y): (x * 100 / n, y * 100 / n)
        transform = lambda (x,y): (x + 0.5*y, sqrt(3)/2 * y)

        for (x1, y1), (x2, y2) in walls:
            x1, y1 = transform(scale((x1, y1)))
            x2, y2 = transform(scale((x2, y2)))
            s.append(svg.SVG('line', x1=x1, y1=y1, x2=x2, y2=y2))

        return svg.canvas(s)
```

```python
TriangleMaze(20).draw().save('img/triangle_maze.svg')
SVG('img/triangle_maze.svg')
```
