# Chaos game

```python
import random
import Image
from math import sin, cos, pi
from IPython.display import Image as iImage

def chaos(points, r, n):
    """ Draw chaos game. """
    points_x, points_y = zip(*points)
    colors = [(255., 0., 0.)[(i % 3):] + (255., 0., 0.)[:(i % 3)]
              for i in range(len(points))]

    size_x, size_y = int(max(points_x)), int(max(points_y))
    im = Image.new('RGB', (size_x, size_y), (255,255,255))

    # random intial point and color
    x = (random.uniform(0, size_x), random.uniform(0, size_y))
    c = (random.uniform(0,255), random.uniform(0,255), random.uniform(0,255))

    affine = lambda a, b, r: [ai * r + bi * (1 - r) for ai, bi in zip(a, b)]

    for _ in range(n):
        chosen, chosen_c = random.sample(zip(points, colors), 1)[0]
        x = affine(x, chosen, r)
        c = affine(c, chosen_c, r)
        try:
            im.putpixel((int(x[0]), int(x[1])), tuple(map(int, c)))
        except IndexError:    # rounding errors
            pass

    return im


def generate_polygon(r=300., n=3):
    points = []
    for i in range(n):
        x, y = r * cos(2*i*pi/n) + r, r * sin(2*i*pi/n) + r
        points.append((x,y))
    return points


def save_and_display(n, r, name):
    points = generate_polygon(200., n)
    im = chaos(points, r, 500000)
    im.save(name)
    return iImage(name)
```

```python
save_and_display(3, 0.5, 'img/triangle.png')
```
