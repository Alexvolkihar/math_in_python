# Fractals

```python
from __future__ import division
from math import sqrt, log
from PIL import Image
import colorsys
from numpy import linspace
log2 = lambda x: log(x, 2)

colscale = lambda c: (int(c[0] * 255), int(c[1] * 255), int(c[2] * 255))

def linspace(start, stop, n):
    if n == 1:
        yield stop
        return
    h = (stop - start) / float(n - 1)
    for i in range(n):
        yield start + h * i

def palette(n=100, rng=3, system='RGB', scaled=False):
    h = map(lambda x: x % 1, linspace(0 + 0.6, 3 + 0.6, n))
    s = [1.] * n
    v = map(lambda x: x**0.2, linspace(0, 1, n))
    if system == 'RGB':
        cols = [colorsys.hsv_to_rgb(*x) for x in zip(h, s, v)]
    else:
        cols = zip(h, s, v)

    if scaled:
        cols = map(colscale, cols)

    return cols


class Fractal(object):

    def draw(self, size=640, center=complex(0,0), d=2):
        """ 
        :param size: image size in pixels
        :param center: center of image as complex number
        :param d: diameter for complex domain
        :return: image object
        """
        im = Image.new('RGB', (size, size))

        colscale = lambda r,g,b: (int(r * 255), int(g * 255), int(b * 255))

        for i, a in enumerate(linspace(center.real - d, center.real + d, size)):
            for j, b in enumerate(linspace(center.imag - d, center.imag + d, size)):
                color = self.complex_color(complex(a,b))
                im.putpixel((i, j), colscale(*color))

        return im

    def complex_color(self, z):
        """ Return color of a pixel for given complex number. """
        raise NotImplementedError
```

## Newton fractal

```python
class Newton(Fractal):
    SOLUTIONS = [1, complex(-0.5, sqrt(3)/2), complex(-0.5, -sqrt(3)/2)]

    def __init__(self, prec=0.001, intensity=20.):
        self.prec = prec
        self.intensity = intensity

    def n_coloring(self, k, iters, intensity):
        return colorsys.hsv_to_rgb(k / 3, 1., max(1. - iters / intensity, 0.))

    def complex_color(self, z):
        i = 0
        while True:
            i += 1
            for k in range(3):
                if abs(z - self.SOLUTIONS[k]) < self.prec:
                    return self.n_coloring(k, i, self.intensity)
            try:
                z = z - (z*z*z - 1) / (3*z*z)
            except ZeroDivisionError:
                return (0,0,0)
```

## Mandelbrot fractal

```python
class Mandelbrot(Fractal):
    def __init__(self, max_iter=50, escape=4., coloring=0):
        self.max_iter = max_iter
        self.escape = escape
        self.coloring = [self.m_coloring_0, self.m_coloring_1, self.m_coloring_2][coloring]

    def m_coloring_2(self, r):
        h, s, v = (3*r % 1, 1., r**0.2)
        return colorsys.hsv_to_rgb(h, s, v)

    def m_coloring_1(self, r):
        if int(r * 100) % 2 == 0:
            return (0,0,0)
        else:
            return (1,1,1)

    def m_coloring_0(self, r):
        return (r**0.6, r**0.4, r**0.2)

    def complex_color(self, c):
        z = 0
        for i in range(self.max_iter):
            z = z * z + c
            if abs(z) > self.escape:
                return self.coloring((i - log2(log2(abs(z)) / log2(self.escape))) / self.max_iter)
        return (0.,0.,0.)
```

## Julia fractal

```python
class Julia(Fractal):
    PALETTE = palette(100, rng=2, system='RGB')

    def __init__(self, c=complex(-0.13,0.75), max_iter=50, escape=2.):
        self.max_iter = max_iter
        self.escape = escape
        self.c = c

    def j_coloring(self, i):
        return self.PALETTE[i % len(self.PALETTE)]

    def complex_color(self, z):
        for i in range(self.max_iter):
            z = z * z + self.c
            if abs(z) > self.escape:
                return self.j_coloring(i+1)
        return (0.,0.,0.)
```
