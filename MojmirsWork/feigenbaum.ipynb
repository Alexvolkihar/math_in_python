# Feigenbaum diagram

```python
import Image
import numpy as np

def fei(r, x0=0.25, iters=500):
    """ Generate 100 x-values for given r from logistic regression. Drop
    first 100 values. """
    ret = [x0]
    for _ in range(iters-1):
        ret.append(4 * r * ret[-1] * (1 - ret[-1]))
    return ret[100:]


def diagram(r_span=(0.,1.), x_span=(0.,1.)):
    """ Return 1000 x 1000 image. """
    size_x = 500
    size_y = 500
    im = Image.new('L', (size_x, size_y), 255)

    norm = lambda x, span: (x - span[0]) / (span[1] - span[0])
    scale = lambda x, size: min(int(round(size * x)), size - 1)
    for r in np.linspace(r_span[0], r_span[1], size_x):
        for x in fei(r):
            if x_span[0] <= x <= x_span[1]:
                im.putpixel((scale(norm(r, r_span), size_x),
                             scale(norm(x, x_span), size_y)), 0)

    return im
```

## Feigenbaum diagram for $r \in [0.7, 1], x \in [0, 1]$

```python
from IPython.display import Image as iImage
im = diagram((0.7,1.))
im.save('img/feiegn_full.png')
iImage('img/feiegn_full.png')
```

## Feigenbaum diagram for $r \in [0.85, 9], x \in [0.3, 6]$
