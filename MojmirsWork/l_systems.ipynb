# L-systems

```python
from turtle import Turtle
from math import pi
from IPython.display import SVG

class Lsystem(object):

    def __init__(self, axiom, rules, interp):
        self.axiom = axiom
        self.rules = rules
        self.interp = interp

        
    def gen_instructions(self, n):
        spam = self.axiom
        for _ in range(n):
            new = ''
            for x in spam:
                new += self.rules.get(x, x)
            spam = new
        return spam


    def draw(self, n, center=(10,70)):
        turtle = Turtle(center)

        instr = self.gen_instructions(n)
        for s in instr:
            if s in self.interp:
                eval('turtle.' + self.interp[s])

        return turtle
```

```python
koch = Lsystem('F--F--F',
                    {'F': 'F+F--F+F'},
                    {'F': 'forward(0.3)', '+': 'right(pi/3)', '-': 'left(pi/3)'}
                    )
koch.draw(5).save('img/koch.svg')
SVG('img/koch.svg')
```

```python
sierp = Lsystem('A',
                {'A': 'B-A-B', 'B': 'A+B+A'},
                {'A': 'forward(0.5)', 'B': 'forward(0.5)', '+': 'right(pi/3)', '-': 'left(pi/3)'}
                )
sierp.draw(7).save('img/l_sierpinsky.svg')
SVG('img/l_sierpinsky.svg')
```

```python
hilbert = Lsystem('A',
                {'A': '-BF+AFA+FB-', 'B': '+AF-BFB-FA+'},
                {'F': 'forward(1.)', '+': 'right(pi/2.)', '-': 'left(pi/2.)'}
                )
hilbert.draw(6, (20,80)).save('img/hilbert.svg')
SVG('img/hilbert.svg')
```

```python
tree = Lsystem('A',
                {'A': 'F[+A]-A', 'F': 'FF'},
                {'F': 'forward(0.5)', '+': 'right(pi/4)', '-': 'left(pi/4)',
                 '[': 'push()', ']': 'pop()'}
                )
tree.draw(7, (20,50)).save('img/tree.svg')
SVG('img/tree.svg')
```

```python
wheat = Lsystem('A',
                {'A': 'F-[[A]+A]+F[+FA]-A', 'F': 'FF'},
                {'F': 'forward(0.6)', '+': 'right(0.14*pi)', '-': 'left(0.14*pi)',
                 '[': 'push()', ']': 'pop()'}
                )
wheat.draw(6, (0,50)).save('img/wheat.svg')
SVG('img/wheat.svg')
```

```python
bush = Lsystem('++++F',
                {'F': 'FF-[-F+F+F]+[+F-F-F]'},
                {'F': 'forward(1.8)', '+': 'right(0.09*pi)', '-': 'left(0.09*pi)',
                 '[': 'push()', ']': 'pop()'}
                )
bush.draw(4, (0,0)).save('bush.svg')
SVG('img/bush.svg')
```

```python
dragon = Lsystem('FX',
                {'X': 'X+YF+', 'Y': '-FX-Y', 'F': ''},
                {'F': 'forward(0.7)', '+': 'right(pi/2)', '-': 'left(pi/2)'}
                )
dragon.draw(14, (30,70)).save('img/dragon.svg')
SVG('img/dragon.svg')
```
